```python

hist,bins = np.histogram((rintin_gray).flatten(),256,[0,256])

cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()

plt.plot(cdf_normalized, color = 'b')
plt.hist(rintin_gray.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```

```python
redWithOutGreen = hall_r-hall_g
```

```python
def pixels_red(a):
    if a > 128 and a < 179: #chute
        return 255
    else:
        return 0

pixels_red_v = np.vectorize(pixels_red)

pixels_box = [pixels_red_v(line) for line in redWithOutGreen]
pixels_box = np.asarray (pixels_box)

plt.imshow (pixels_box, cmap = "Greys_r")
```
